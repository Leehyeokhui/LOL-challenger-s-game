### 상대보다 더 많은 드래곤을 처치했을 때의 승률은 어떨까?

```python
# blue 기준으로 계산
df3 = df[['blueWins', 'redWins', 'blueDragonKills', 'redDragonKills']]
df3['킬 차이'] = (df3['blueDragonKills'] - df3['redDragonKills'])
df3
```

```python
df_md = df3[df3['킬 차이'] > 0] # blue가 더 많은 드래곤을 처치했을 경우
df_md
```

```python
# 약 83%의 승률
sum(df_md['blueWins'])/len(df_md)
```

```python
df_md2 = df3[df3['킬 차이'] < 0] # red가 더 많은 드래곤을 처치했을 경우
df_md2
```

```python
# 약 80%의 승률
sum(df_md2['redWins'])/len(df_md2) 
```

더 많은 오브젝트는 승리의 중요한 포인트 - 너무 당연한 이야기....
